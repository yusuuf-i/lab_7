Bu notebook, `teleCust1000t.csv` veri setini kullanarak K-Nearest Neighbors (KNN) algoritması ile müşteri segmentasyonu yapmaktadır.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix, classification_report
```

### 1. Veriyi Yükleme ve İnceleme
**ÖNEMLİ NOT:** Colab kullanıyorsanız `teleCust1000t.csv` dosyasını sol taraftaki 'Dosyalar' paneline yüklemeniz gerekmektedir.

```python
df = pd.read_csv('teleCust1000t.csv')
print("--- Veri Setinin İlk 5 Satırı ---")
display(df.head())

print("\n--- Sınıf Dağılımı (Müşteri Kategorileri) ---")
print(df['custcat'].value_counts())
```

### 2. Veri Hazırlığı ve Normalizasyon
Hedef değişkeni ayırıp eğitim ve test veri seti olarak ikiye bölüyoruz. Ayrıca KNN algortiması uzaklık tabanlı çalıştığından değişkenlerimizi StandardScaler ile normalize ediyoruz.

```python
X = df.drop('custcat', axis=1).values
y = df['custcat'].values

scaler = StandardScaler()
X_normalized = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_normalized, y, test_size=0.2, random_state=42)

print("Eğitim Seti (X_train):", X_train.shape)
print("Test Seti (X_test):", X_test.shape)
```

### 3. K=4 ve K=6 İçin Model Eğitimi ve Tahmin
Önce k=4 sonra k=6 değerleri için model eğitilip performans raporları üretiliyor.

```python
# K=4 için
knn_4 = KNeighborsClassifier(n_neighbors=4)
knn_4.fit(X_train, y_train)
y_pred_4 = knn_4.predict(X_test)
print("--- K=4 İçin Performans Sonuçları ---")
print(f"Accuracy Score: {accuracy_score(y_test, y_pred_4):.4f}")
print(classification_report(y_test, y_pred_4))

# K=6 için
knn_6 = KNeighborsClassifier(n_neighbors=6)
knn_6.fit(X_train, y_train)
y_pred_6 = knn_6.predict(X_test)
print("--- K=6 İçin Performans Sonuçları ---")
print(f"Accuracy Score: {accuracy_score(y_test, y_pred_6):.4f}")
print(classification_report(y_test, y_pred_6))
```

### 4. Optimum K Değerinin Bulunması
En iyi performansı veren K değerini bulabilmek için 1'den 10'a kadar tüm K değerlerini deneyip test accuracy değerlerini grafik olarak bastırıyoruz.

```python
k_values = range(1, 11)
accuracy_scores = []

for i in k_values:
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    accuracy_scores.append(accuracy_score(y_test, y_pred))

print("\n--- 1'den 10'a Kadar K Değerleri İçin Doğruluk Skorları ---")
for i, score in enumerate(accuracy_scores, 1):
    print(f"K={i} -> Accuracy: {score:.4f}")

plt.figure(figsize=(10, 6))
plt.plot(k_values, accuracy_scores, marker='o', linestyle='dashed', color='blue', markersize=8, markerfacecolor='red')
plt.title('K Değerine Göre Doğruluk (Accuracy) Grafiği')
plt.xlabel('K Değeri (Komşu Sayısı)')
plt.ylabel('Doğruluk Oranı (Accuracy)')
plt.xticks(k_values)
plt.grid(True)
plt.show()
```
